```python
# 1
med_bairro = dados.groupby(["Bairro"])["Metragem"].mean()
med_bairro
```

```python
# 2.1
dados.iloc[:,[1]]
```

```python
# 2.2
bairro = "Mooca"
loc_bairro = dados.loc[dados['Bairro'] == bairro]
loc_bairro
```

```python
# 3 
# Média de quartos por bairro
quartos = dados.groupby('Bairro')['Quartos'].mean()
quartos.head(10).plot.bar(ylabel="Média de quartos")
```

```python
# Número de imóveis por bairro
n_imoveis_bairro = dados["Bairro"].value_counts()
n_imoveis_bairro.head(10).plot.barh(color='red')
```
